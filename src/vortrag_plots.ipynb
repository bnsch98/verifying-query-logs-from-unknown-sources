```python
## create dummy data bar plot with 3 bars
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


# Create dummy data
data = {
    'Category': ['A', 'B', 'C'],
    'Value': [1/6, 3/6, 2/6]
}
df = pd.DataFrame(data)
# Set the figure size
plt.figure(figsize=(8, 6))
# Create a bar plot
# blue color
# green color map
# Set the color palette
ax = sns.barplot(x='Category', y='Value', data=df, palette='viridis') #'viridis'
# Set the title and labels
# plt.title('Query Intent')
# plt.xlabel('Category')
# plt.ylabel('Value')
# Set the y-axis limits
plt.ylim(0, 0.6)
# Add grid lines
# ax.yaxis.grid(True, linestyle='--', alpha=0.7)
# Customize the y-axis ticks
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
# Customize the x-axis ticks
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
# Rotate x-axis labels
plt.xticks(rotation=45)
# dont show x label and y label
plt.xlabel('')
plt.ylabel('')
# Show the plot
plt.tight_layout()
plt.show()
# save plot as pdf
outputpath = '/home/benjamin/studium/masterarbeit/presentation/test-sigir/figures/'
# plt.savefig(outputpath+'bar-plot-aql.pdf', bbox_inches='tight')
```

```python
# make dummy data for plotting 3 time series

# Create dummy data
time = np.arange(0, 10, 1)

# the time series data should simulate a temporal popularity of a query
data1 = -np.sin(time) + np.random.normal(0, 0.1, len(time))
data2 = -np.sin(time + 1) + np.random.normal(0, 0.1, len(time))
data3 = -np.sin(time + 2) + np.random.normal(0, 0.1, len(time))
# Create a DataFrame
df = pd.DataFrame({
    'Time': time,
    'Data1': data1,
    'Data2': data2,
    'Data3': data3,
})
# Set the figure size
plt.figure(figsize=(8, 6))
# Create a line plot
ax = sns.lineplot(x='Time', y='Data1', data=df, label='Query 1')
ax = sns.lineplot(x='Time', y='Data2', data=df, label='Query 2')
ax = sns.lineplot(x='Time', y='Data3', data=df, label='Query 3')

# Set the title and labels
# plt.title('Time Series Data')
# plt.xlabel('Time')
# plt.ylabel('Value')
# Set the y-axis limits
plt.ylim(-2, 2)
# Add grid lines
# ax.yaxis.grid(True, linestyle='--', alpha=0.7)
# Customize the y-axis ticks
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
# Customize the x-axis ticks
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
# Rotate x-axis labels
plt.xticks(rotation=45)
# dont show x label and y label
plt.xlabel('Month')
plt.ylabel('Frequency')
# Show the plot
plt.tight_layout()
# plt.show()
# save plot as pdf
outputpath = '/home/benjamin/studium/masterarbeit/presentation/test-sigir/figures/'
plt.savefig(outputpath+'time-series-aql.pdf', bbox_inches='tight')
```

```python
# nnext we plot a log log plot of a zipfs law
# we take 3 dummy data sets and plot them in a log log plot 

# Create dummy data
x = np.arange(1, 1000)
y1 = 1 / (x ** 1.5 +0.1)  # Zipf's law
y2 = 1 / (x ** 1.7)  # Zipf's law
y3 = 1 / (x ** 1.9)  # Zipf's law

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y1': y1,
    'y2': y2,
    'y3': y3,
})
# Set the figure size
plt.figure(figsize=(8, 6))
# Create a log-log plot
ax = sns.lineplot(x='x', y='y1', data=df)
ax = sns.lineplot(x='x', y='y2', data=df)
ax = sns.lineplot(x='x', y='y3', data=df)
# Set the title and labels
# plt.title('Zipf\'s Law')
# plt.xlabel('Rank')
# plt.ylabel('Frequency')
# Set the y-axis limits 
plt.ylim(0, 1)
# Add grid lines
# ax.yaxis.grid(True, linestyle='--', alpha=0.7)
# Customize the y-axis ticks
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
# Customize the x-axis ticks
ax.xaxis.set_major_locator(ticker.MultipleLocator(100))
# Rotate x-axis labels
plt.xticks(rotation=45)
# dont show x label and y label
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.grid(which='both', linestyle='--', alpha=0.7)

# log scale
plt.xscale('log')
plt.yscale('log')
# Set the x-axis limits
plt.xlim(1, 100)
# Set the y-axis limits
plt.ylim(0.001, 1)

# Customize the x-axis ticks

# Show the plot
plt.tight_layout()
# plt.show()
# save plot as pdf
outputpath = '/home/benjamin/studium/masterarbeit/presentation/test-sigir/figures/'
plt.savefig(outputpath+'rank-size-plot-truth.pdf', bbox_inches='tight')
```
